```python
import selenium
import time
import csv
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

PATH = Service("C:\Program Files (x86)\chromedriver.exe")
```

```python
################
#### PART 1 ####
################

driver = webdriver.Chrome(service=PATH)

# Navigate to a website
driver.get('https://techwithtim.net')

# Finds element by link text and clicks link
link = driver.find_element(By.LINK_TEXT, "Python Programming")
link.click()

# Makes sure that the element is loaded on the page before trying to interact
# with it. Gives the page 10 seconds to load before continuing
try:
    # Finds first element
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.LINK_TEXT, "Beginner Python Tutorials"))
    )
    element.click()
    
    # Finds second element
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "sow-button-19310003"))
    )
    element.click()
    
    # Brings us back to homepage using driver.back()
    driver.back()
    driver.back()
    driver.back()
    # driver.forward() would bring us to the forward page
    

except:
    driver.quit()
```

```python
################
#### PART 2 ####
################

# Set up csv file and driver
csv_file = open("players_salaries.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Player", "Salary"])

driver = webdriver.Chrome(service=PATH)

# Navigate to the website
driver.get('https://hoopshype.com/')

# Makes sure that the element is loaded on the page before trying to interact
# with it. Gives the page 10 seconds to load before continuing
try:
    # Finds first element (salaries)
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.LINK_TEXT,"SALARIES"))
    )
    element.click()
    
    # Finds second element (player salaries)
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.LINK_TEXT, "PLAYER SALARIES"))
    )
    element.click()
        
    # Gets all items in table rows. This includes the players name and salaries for the next 3 years, formatted as:
    # " [rank] [player name] [2022-23 salary] [2023-2024 salary] [2024-25 salary] ""
    # all as one string when converted to text.
    players_salaries = driver.find_elements(By.CSS_SELECTOR, "tr")
    players_salaries.remove(players_salaries[0])
    
    # uses string maniupulation to extract just the player name and current salary of the player, then writes to csv
    for i in players_salaries:
        data = i.text   # turns the current element into a string
        parts = data.split()  # split the string into words
        name = " ".join(parts[1:-4])  # extract the name
        salary = parts[3]  # extract the first salary number
        
        csv_writer.writerow([name, salary])

except Exception:
    driver.quit()
    print(Exception)
    
csv_file.close()
```
